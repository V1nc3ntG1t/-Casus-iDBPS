<h3>Distribution of the Data - normality test

```python
from scipy.stats import ttest_ind

x = df_final.copy()

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64', 'uint8']
xNumerics = df_final.select_dtypes(include=numerics)
#Confirm assumtion: Observations in each sample are independent and identically distributed (iid).

#Because of the small dataset there are some columns that always have the same value, these columns are removed for the normality test (identical)
xIndependant = xNumerics
for i in xIndependant:
    a = xIndependant[i].unique()
    if len(a) <= 1:
        xIndependant = xIndependant.drop(columns=[i])
xIndependant.columns
```

```python
#Shapiro-Wilk Test
from scipy.stats import shapiro
    
for i in xIndependant:
    stat, p = shapiro(xIndependant[i])
    if p > 0.05:
        #print columns that are probably Gaussian
        print(i)
```

```python
#D’Agostino’s K^2 Test
from scipy.stats import normaltest

for i in xIndependant:
    stat, p = normaltest(xIndependant[i])
    if p > 0.05:
        #print columns that are probably Gaussian
        print(i)
```

```python
#Anderson-Darling Test
from scipy.stats import anderson


for i in xIndependant:
    result = anderson(xIndependant[i])
    for j in range(len(result.critical_values)):
        sl, cv = result.significance_level[j], result.critical_values[j]
        if result.statistic < cv:
            #print columns that are probably Gaussian
            print(i)
```

<h2>Non Parametrical Hypothesis Test

```python
y = x[['Respiratory Rate (1/min)', 'Ventilation (sL/min)']].copy()

#first we check if  the columns are dependent
listI =[]
for i in y:
    for j in y:
        if (i != j) & (j not in listI):
            print('\n',i, j)
            stat, p = spearmanr(y[i], y[j])
            if p > 0.05:
                print('Probably independent')
            else:
                print('Probably dependent')
    listI.append(i)
```

```python
# the Wilcoxon Signed-Rank Test
from scipy.stats import wilcoxon

listI =[]
for i in y:    
    for j in y:
        if (i != j) & (j not in listI):
            print('\n',i, j)
            stat, p = wilcoxon(y[i],y[j])
            if p > 0.05:
                print('Probably the same distribution')
            else:
                print('Probably different distribution')
    listI.append(i)
```

<h4> VARIATION INFLATION FACTOR (VIF)

```python
df_vif = x
```

```python
import time
from joblib import Parallel, delayed

# Defining the function that you will run later
#https://github.com/K-Ellis/Neural-Network-Classification-and-Unbalanced-Bank-Marketing-Data/issues/1#issue-234132765
def calculate_vif_(X, thresh=5.0):
    variables = [X.columns[i] for i in range(X.shape[1])]
    dropped=True
    while dropped:
        dropped=False
        print(len(variables))
        vif = Parallel(n_jobs=-1,verbose=5)(delayed(variance_inflation_factor)(X[variables].values, ix) for ix in range(len(variables)))

        maxloc = vif.index(max(vif))
        if max(vif) > thresh:
            print(time.ctime() + ' dropping \'' + X[variables].columns[maxloc] + '\' at index: ' + str(maxloc))
            variables.pop(maxloc)
            dropped=True

    print('Remaining variables:')
    print(variables)
    
    return X[[i for i in variables]]

#getting the VIF score

def vif_scores(df):
    VIF_Scores = pd.DataFrame()
    VIF_Scores["Independent Features"] = df.columns
    VIF_Scores["VIF Scores"] = [variance_inflation_factor(df.values,i) for i in range(df.shape[1])]
    return VIF_Scores
```

```python
#vif scores on all columns
vif_scores(df_vif)
```
